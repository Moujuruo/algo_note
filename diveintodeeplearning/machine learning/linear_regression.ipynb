```python
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
x = np.array([56, 72, 69, 88, 102, 86, 76, 79, 94, 74])
y = np.array([92, 102, 86, 110, 130, 99, 96, 102, 105, 92])
```

```python
from matplotlib import pyplot as plt

%matplotlib inline

plt.scatter(x, y)
plt.xlabel("Area")
plt.ylabel("Price")
```

```python
def square_loss(y_hat, y):
    return (y_hat - y.reshape(y_hat.shape))**2 / 2

def least_squares_matrix(x, y):
    return (x.T * x).I * x.T * y
```

```python
x_matrix = np.mat(np.concatenate((np.ones((x.shape[0], 1)), x.reshape(-1, 1)), axis=1))
y_matrix = np.mat(y.reshape(-1, 1))

w = least_squares_matrix(x_matrix, y_matrix)
```

```python
x_tmp = np.arange(50, 110, 0.1)
y_tmp = w[0] + w[1] * x_tmp

plt.scatter(x, y)
plt.plot(x_tmp, y_tmp.reshape(-1, 1), color='red')
```
