```python
def classify(data, features):
    # data 是训练数据, features 是待分类的特征
    labels = data["labels"]
    # 计算先验概率, P(Y=ck)
    labels_p = labels.value_counts() / len(labels)
    labels_p = dict(labels_p)
    # 计算条件概率, P(X=xi|Y=ck)
    features_p = {}
    for feature in features:
        for label in labels.unique():
            text = feature + "|" + label
            data_label = data[data["labels"] == label]
            # feature_p = [data[i] for i, item in data_label.iterrows() if item['x'] == feature or item['y'] == feature]
            feature_p = [item for i, item in data_label.iterrows() if item['x'] == feature or item['y'] == feature]

            feature_p = len(feature_p) / len(data_label)
            features_p[text] = feature_p

    # 计算后验概率, P(Y=ck,X) = P(X|Y=ck)P(Y=ck)
    posterior_p = {}
    for label in labels.unique():
        posterior_p[label] = labels_p[label]
        for feature in features:
            text = feature + "|" + label
            posterior_p[label] *= features_p[text]

    # print(posterior_p)
    return max(posterior_p, key=posterior_p.get), posterior_p

```

```python
classify(data, ['r', 'm'])
```
